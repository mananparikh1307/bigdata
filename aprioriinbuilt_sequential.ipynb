```python
import numpy as np
import pandas as pd
from apyori import apriori
import timeit
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
file = "/content/gdrive/My Drive/mushrooms.csv"
store_data  = pd.read_csv(file,header = None)
```

```python
store_data
```

```python
store_data.shape
```

```python
start = timeit.default_timer()
records = []
for i in range(1,8125):
  records.append([str(store_data.values[i,j]) for j in range(1,23)])
association_rules = apriori(records,min_support = 0.5 ,  min_confidence = 1 , min_lift =0 ,min_length =1  )
association_results = list(association_rules)
print(association_results)
stop = timeit.default_timer()

print('Time: ', stop - start) 
```
